```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as syp
```

```python
sentiment_csv = pd.read_csv("Files/apple_sentiments.csv")
price_csv = pd.read_csv("Files/AAPL.csv")
```

```python
# Grouping table to find dates
grouped_by_date = sentiment_csv.groupby("Date")
date = list(grouped_by_date.groups.keys())

#Company Name for each row
company_name = []
for x in range(len(date)):
     company_name.append(sentiment_csv["Company"][x])

#Average SEntiment
avr_sentiment = list(grouped_by_date["Sentiment"].mean())

#Number of articles
num_articles = list(grouped_by_date["Headline"].count())


#NEgative Counts
neg_counts = sentiment_csv.loc[sentiment_csv["Sentiment"]<0].groupby("Date").count()
neg_news_count = list(neg_counts["Sentiment"])

#Positive Counts

pos_counts = sentiment_csv.loc[sentiment_csv["Sentiment"]>0].groupby("Date").count()
pos_news_count = list(pos_counts["Sentiment"])

#Neutral Counts
neut_counts =[num_articles[x]-(pos_news_count[x]+neg_news_count[x]) for x in range(len(date))]

#Change on Day
cod_price = price_csv["Percent Change"]
```

```python
master_df = pd.DataFrame({"Date":date, "Company": company_name, "Num_aticles":num_articles,"Neg_article_count":neg_news_count, "Pos_article_count":pos_news_count,"Neut_article_count": neut_counts, "COD_Price":cod_price, "Avr sentiment": avr_sentiment})
master_df.head()
```
