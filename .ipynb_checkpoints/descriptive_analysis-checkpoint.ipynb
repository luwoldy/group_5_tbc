```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as syp
import statistics
import glob
import csv
```

```python
# Combining Stock Price CSV files into one file as ALL_Prices.csv

headers=["ID", "Date", "Percent Change"]
csvfiles=glob.glob(r"Stock Prices/*.csv")       
wf=csv.writer(open('Files/All_Prices.csv', 'w'),delimiter=',', lineterminator='\n')
wf.writerow(dict((fn,fn) for fn in headers))

for file in csvfiles:
    rd=csv.reader(open(file,'r'), delimiter=',')
    next(rd)
    for row in rd:
        print(row)
        wf.writerow(row)
```

```python
# Combining Sentiments CSV files into one file as All_Sentiments.csv
headers=["ID", "Date", "Sentiment", "Company"]
csvfiles=glob.glob("Sentiments/*.csv")       
wf=csv.writer(open('Files/All_Sentiments.csv', 'w'),delimiter=',', lineterminator='\n')
wf.writerow(dict((fn,fn) for fn in headers))

#+++ Modify if new stock is added
names=["Amazon", "Apple", "Bank of America", "Exxon", "Verizon"]
for file in csvfiles:
#+++ add a new elif condition new stock file is added       
    if file=="Sentiments\\Amazon_sentiments_final.csv":
        company=names[0]
    elif file=="Sentiments\\apple_sentiments_final.csv":
        company=names[1]
    elif file=="Sentiments\\BofA__sentiments_final.csv":
        company=names[2]
    elif file=="Sentiments\\exxon_final.csv":
        company=names[3]
    elif file=="Sentiments\\verizon_final.csv":
        company=names[4]
    rd=csv.reader(open(file,'r'), delimiter=',')
    next(rd)
#Looping and Writing each line to the csv file
    for row in rd:
        print(row+[company])
        wf.writerow(row+[company])
```
