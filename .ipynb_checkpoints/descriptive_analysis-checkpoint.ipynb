```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as syp
import statistics
import glob
import csv
```

```python
# Combining Stock Price CSV files into one file as ALL_Prices.csv

headers=["ID", "Date", "Percent Change"]
csvfiles=glob.glob(r"Stock Prices/*.csv")       
wf=csv.writer(open('Files/All_Prices.csv', 'w'),delimiter=',', lineterminator='\n')
wf.writerow(dict((fn,fn) for fn in headers))

for file in csvfiles:
    rd=csv.reader(open(file,'r'), delimiter=',')
    next(rd)
    for row in rd:
        print(row)
        wf.writerow(row)
```

```python
# Combining Sentiments CSV files into one file as All_Sentiments.csv
headers=["ID", "Date", "Sentiment", "Company"]
csvfiles=glob.glob("Sentiments/*.csv")       
wf=csv.writer(open('Files/All_Sentiments.csv', 'w'),delimiter=',', lineterminator='\n')
wf.writerow(dict((fn,fn) for fn in headers))

#+++ Modify if new stock is added
names=["Amazon", "Apple", "Bank of America", "Exxon", "Verizon"]
for file in csvfiles:
#+++ add a new elif condition new stock file is added       
    if file=="Sentiments\\Amazon_sentiments_final.csv":
        company=names[0]
    elif file=="Sentiments\\apple_sentiments_final.csv":
        company=names[1]
    elif file=="Sentiments\\BofA__sentiments_final.csv":
        company=names[2]
    elif file=="Sentiments\\exxon_final.csv":
        company=names[3]
    elif file=="Sentiments\\verizon_final.csv":
        company=names[4]
    rd=csv.reader(open(file,'r'), delimiter=',')
    next(rd)
#Looping and Writing each line to the csv file
    for row in rd:
        print(row+[company])
        wf.writerow(row+[company])
```

```python
#Reading data from combined csv files
sentiment_csv = pd.read_csv("Files/All_Sentiments.csv")
price_csv = pd.read_csv("Files/All_Prices.csv")
```

```python
# Grouping table to find dates
grouped_by_date = sentiment_csv.groupby(["Company","Date"])
company_and_date = list(grouped_by_date.groups.keys())
date = [company_date[x][1] for x in range(len(company_and_date))]

index = [x for x in range(len(company_date))]
name=[company_date[x][0] for x in range(len(company_and_date))]

# #Average SEntiment
avr_sentiment =list(grouped_by_date["Sentiment"].mean())


# # #Number of articles
num_articles = list(grouped_by_date["Sentiment"].count())

#NEgative Counts
neg_counts = sentiment_csv.loc[sentiment_csv["Sentiment"]<0].groupby(["Company","Date"]).count()
neg_news_count = list(neg_counts["Sentiment"])
neg_news_percent = [int(neg_news_count[x]/num_articles[x]*-100) for x in range(0,len(date))]

#Positive Counts
pos_counts = sentiment_csv.loc[sentiment_csv["Sentiment"]>0].groupby(["Company","Date"]).count()
pos_news_count = list(pos_counts["Sentiment"])
pos_news_percent = [int(pos_news_count[x]/num_articles[x]*100) for x in range(0,len(date))]

#Neutral Counts
neut_counts =[num_articles[x]-(pos_news_count[x]+neg_news_count[x]) for x in range(len(date))]
neut_news_percent = [int((neut_counts[x]/num_articles[x])*100) for x in range(0,len(date))]

cod_price = price_csv["Percent Change"]
```

```python
#Creating master Data frame to combine data sets
master_df = pd.DataFrame({"Date":date, "Company":name,"Num_aticles":num_articles,"Neg_news_percent":neg_news_percent, "Pos_news_percent":pos_news_percent,"Neut_article_percent": neut_news_percent, "COD_Price":cod_price, "Avr_sentiment": avr_sentiment})
master_df["Positive"]=  master_df['Avr_sentiment'] > 0
master_df.head()
```

```python
# Describtive analysis is needed to check data accuracy and number of observations
master_df.describe()
```

```python
%matplotlib notebook
fig1 = plt.figure(figsize=(17, 10),dpi=60)

plt.bar(index, master_df["Avr_sentiment"], alpha=0.7, align="center",  color=master_df.Positive.map({True: 'g', False: 'r'}))

plt.plot( master_df["COD_Price"], color='b', marker="*", label="C", lw='2')

plt.tick_params(axis='x', which='major',  labelsize=16)

plt.title("Stock News Sentiment", fontsize=21)
plt.xlabel("Daily Range", fontsize=18)
plt.ylabel("Price % Change & Sentiment", fontsize=18)

plt.grid(alpha=0.5)
plt.show()
```

```python
%matplotlib notebook
fig2 = plt.figure(figsize=(12, 8),dpi=60)
plt.scatter(master_df["Avr_sentiment"],master_df["COD_Price"], edgecolors="black", alpha =0.8, linewidths=1)

plt.xlabel("Average Sentiment")
plt.ylabel("Stock Price Change")
plt.title("Stock Price change vs Average News Sentiment", fontsize=20)
plt.grid(alpha=0.4)
```

```python
grouped_by_company = master_df.groupby("Company")
grouped_df = grouped_by_company.mean()
grouped_df
```
